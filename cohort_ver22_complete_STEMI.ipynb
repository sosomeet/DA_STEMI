### 3.2.1 결측 데이터 누락 경로 확인

```python
import pandas as pd
import os
import sqlite3
import numpy as np

base_path = '/content/drive/MyDrive/Colab Notebooks/2025-2 데이터애널리틱스'

# [1] 경로 설정

# ED 데이터셋 폴더 경로
ed_base_path = '/content/drive/MyDrive/Colab Notebooks/2025-2 데이터애널리틱스/mimic-iv-ed-2.2/'

# HOSP/ICU 데이터베이스 파일 경로
db_path = '/content/drive/MyDrive/Colab Notebooks/2025-2 데이터애널리틱스/MIMIC4-hosp-icu.db'

# [2] 데이터 로드 및 코호트 생성

# 2-1. ED 데이터 로드 (CSV)
edstays_csv_path = os.path.join(ed_base_path, 'edstays.csv')
df_edstays = pd.read_csv(edstays_csv_path,
                          dtype={'subject_id': int, 'hadm_id': 'Int64', 'stay_id': int})
print(f"edstays.csv 로드 완료 (Shape: {df_edstays.shape})")

# 2-2. HOSP 데이터 로드 (SQLite)
conn = sqlite3.connect(db_path) # DB 연결
query = "SELECT subject_id, hadm_id, icd_code, icd_version FROM diagnoses_icd"
df_diagnoses_icd = pd.read_sql_query(query, conn)
conn.close() # DB 연결 해제

df_diagnoses_icd['icd_code'] = df_diagnoses_icd['icd_code'].astype(str)
print(f"diagnoses_icd DB 테이블 로드 완료 (Shape: {df_diagnoses_icd.shape})")

# [2] ICU 데이터 원본 (icustays) 확인

try:
    conn = sqlite3.connect(db_path) # DB 연결

    # icustays 테이블은 ICU 입/퇴실 시간의 원본 데이터
    # subject_id, hadm_id, icu_intime, icu_outtime을 가져옴
    query_icu = """
    SELECT
        subject_id, hadm_id, stay_id as icu_stay_id, intime as icu_intime_db, outtime as icu_outtime_db
    FROM icustays
    """
    df_icustays = pd.read_sql_query(query_icu, conn)
    conn.close() # DB 연결 해제

    print(f"icustays DB 테이블 로드 완료 (Shape: {df_icustays.shape})")

    # 3. 현재 코호트와 icustays 테이블 병합 시도 (hadm_id 기준)
    # 현재 코호트에 누락된 ICU 기록이 있는지 확인하기 위함
    # hadm_id는 실수형이므로 형변환 (Int64)
    cohort_ver21_hadm['hadm_id'] = cohort_ver21_hadm['hadm_id'].astype('Int64')

    # 1:N 관계이므로, 하나의 hadm_id에 여러 icustay가 있을 수 있음
    df_check = pd.merge(
        cohort_ver21_hadm[['hadm_id', 'icu_intime']].dropna(subset=['hadm_id']),
        df_icustays[['hadm_id', 'icu_stay_id']].drop_duplicates(),
        on='hadm_id',
        how='left',
        indicator=True # 병합 결과 표시
    )

    # 4. ICU 기록이 누락된 행 확인
    # 현재 코호트에는 hadm_id가 있지만, icustays에는 없는 경우를 찾기
    # 'both'가 아닌 'left_only'이면서 icu_intime이 결측인 행의 수를 확인해야 함

    # 현재 코호트에서 icu_intime이 NaN인 행 (465개)을 추출
    df_missing_icu = cohort_ver21_hadm[cohort_ver21_hadm['icu_intime'].isnull()].copy()

    # icu_intime이 NaN인 행들의 hadm_id를 icustays 테이블에서 찾기
    missing_hadm_ids = df_missing_icu['hadm_id'].unique()

    # 이 hadm_id들이 icustays에 존재하는지 확인
    hadm_id_found_in_icu = df_icustays[df_icustays['hadm_id'].isin(missing_hadm_ids)]['hadm_id'].unique()


    print("\n--- ICU 결측치 병합 오류 검증 결과 ---")
    print(f"ICU 기록이 결측인 hadm_id 수 (현재 코호트 기준): {len(missing_hadm_ids)}개")
    print(f"ICU 기록이 결측이었지만 icustays 원본 테이블에서 발견된 hadm_id 수: {len(hadm_id_found_in_icu)}개")

    if len(hadm_id_found_in_icu) == 0:
        print("\n✅ 확인: ICU 기록이 결측인 모든 환자는 원본 icustays 테이블에도 기록이 없습니다.")
        print("결측치 465건은 병합 오류가 아닌 'ICU 미입원'을 의미할 가능성이 매우 높습니다.")
    else:
        print(f"\n⚠️ 경고: {len(hadm_id_found_in_icu)}개의 hadm_id가 ICU 결측 행이었으나, 원본 icustays 테이블에서 발견되었습니다.")
        print("이는 코호트 생성 또는 ICU 데이터 병합 과정에서 문제가 있었음을 의미합니다.")

# [5] 오류 처리 (DB 관련)
except pd.io.sql.DatabaseError as e:
    print(f"\n--- DB 연결 오류 (SQLite) ---")
    print(f"오류: DB 연결이나 'icustays' 테이블 쿼리에 실패했습니다.")
    print(f"경로: '{db_path}'")
    print(f"상세 오류: {e}")

except Exception as e:
    print(f"--- 알 수 없는 오류 ---")
    print(f"오류 발생: {e}")
```

### 3.2.2 ICU 입/퇴원 시간 결측치 보완

```python
import pandas as pd
import numpy as np

# 1. 병합을 위해 hadm_id 타입 통일 (Int64)
cohort_ver21_hadm['hadm_id'] = cohort_ver21_hadm['hadm_id'].astype('Int64')
df_icustays['hadm_id'] = df_icustays['hadm_id'].astype('Int64')

# 2. 중복 방지를 위해 환자별 '첫 번째' ICU 입원 기록만 추출
# 한 번의 입원(hadm_id) 중에 ICU를 여러 번 갔다 왔을 수 있으므로, 가장 먼저 들어간 시간(min)을 가져옴
icustays_first = df_icustays.sort_values('icu_intime_db').groupby('hadm_id').first().reset_index()

# 3. 코호트 데이터와 원본 ICU 데이터 병합 (Left Join)
# suffixes를 사용하여 기존 결측 컬럼과 구분
df_merged = pd.merge(
    cohort_ver21_hadm,
    icustays_first[['hadm_id', 'icu_intime_db', 'icu_outtime_db']], # 필요한 컬럼만
    on='hadm_id',
    how='left'
)

# 4. 결측치 채우기 (Data Imputation)
# 기존 icu_intime이 비어있다면(_original이 NaN), DB에서 가져온 값(_db)으로 채움
# 만약 기존 값도 있고 DB 값도 있다면, 기존 값을 우선합니다 (fillna는 결측일 때만 채움)

# 컬럼이 datetime 형식인지 확실히 하기 위해 변환
for col in ['icu_intime', 'icu_outtime']:
    df_merged[col] = pd.to_datetime(df_merged[col], errors='coerce')

df_merged['icu_intime_db'] = pd.to_datetime(df_merged['icu_intime_db'], errors='coerce')
df_merged['icu_outtime_db'] = pd.to_datetime(df_merged['icu_outtime_db'], errors='coerce')

# 결측치 채우기 (fillna)
df_merged['icu_intime'] = df_merged['icu_intime'].fillna(df_merged['icu_intime_db'])
df_merged['icu_outtime'] = df_merged['icu_outtime'].fillna(df_merged['icu_outtime_db'])

# 임시로 가져온 _db 컬럼 제거
df_merged = df_merged.drop(columns=['icu_intime_db', 'icu_outtime_db'])

# 5. 파생 변수 재계산 (is_icu_stay, icu_los_days)
# 이제 313명의 데이터가 채워졌으므로 플래그와 재원일수를 다시 계산
# ICU 입원 여부 (값이 있으면 1, 여전히 없으면 0)
df_merged['is_icu_stay'] = df_merged['icu_intime'].notna().astype(int)

# ICU 재원 일수 (분 단위 -> 일 단위)
df_merged['icu_los_days'] = (df_merged['icu_outtime'] - df_merged['icu_intime']).dt.total_seconds() / (60 * 60 * 24)

# 6. 결과 확인
print("--- ICU 데이터 복구 및 재계산 완료 ---")
print(f"총 행 수: {len(df_merged)}")

# 복구 후 결측치 확인
remaining_missing = df_merged['icu_intime'].isnull().sum()
print(f"복구 후 남은 ICU 결측치 수 (진짜 미입원): {remaining_missing}개")
print(f"복구된 데이터 수: {464 - remaining_missing}개 (예상치: 313개)")

# ICU 입원 여부 분포
print("\n최종 ICU 입원 여부(is_icu_stay) 분포:")
print(df_merged['is_icu_stay'].value_counts())

# 7. 변수명 업데이트 및 저장
cohort_ver22 = df_merged.copy()
print(cohort_ver22.columns)
```

## 3.2.3 is_icu_stay = 0 인 151명의 death_flag 분석

### (1) 재원일수 확인 및 보완  
1. icu_los_days = 0 으로 채우기 (icu에 재원하지 않았으므로)  
2. los_hospital_days 재계산
3. 중증도를 파악하기 위해 전체 입원 기간을 확인하고, ICU 그룹과 비교

```python
import pandas as pd
import sqlite3
import numpy as np

db_path = '/content/drive/MyDrive/Colab Notebooks/2025-2 데이터애널리틱스/MIMIC4-hosp-icu.db'

# Admissions 테이블 로드
# admittime dischtime deathtime 컬럼 로드
try:
    conn = sqlite3.connect(db_path)
    query = "SELECT hadm_id, admittime, dischtime, deathtime FROM admissions"
    df_admissions = pd.read_sql_query(query, conn)
    conn.close()
    print(f"admissions 테이블 로드 완료 Shape {df_admissions.shape}")
except Exception as e:
    print(f"DB 연결 실패 {e}")

cohort_ver21_hadm = cohort_ver21_hadm.copy()

# is_icu_stay 컬럼 재생성
if 'icu_intime' in cohort_ver21_hadm.columns:
    cohort_ver21_hadm['icu_intime'] = pd.to_datetime(cohort_ver21_hadm['icu_intime'], errors='coerce')
    cohort_ver21_hadm['is_icu_stay'] = cohort_ver21_hadm['icu_intime'].notna().astype(int)
    print("is_icu_stay 컬럼 재생성 완료")
else:
    print("icu_intime 컬럼 부재로 is_icu_stay 생성 불가")

# hadm_id 타입 통일
cohort_ver21_hadm['hadm_id'] = cohort_ver21_hadm['hadm_id'].astype('Int64')
df_admissions['hadm_id'] = df_admissions['hadm_id'].astype('Int64')

# 기존 컬럼 중복 방지를 위한 drop 후 merge
cols_to_drop = ['admittime', 'dischtime', 'deathtime', 'los_days']
cohort_ver21_hadm = cohort_ver21_hadm.drop(columns=[c for c in cols_to_drop if c in cohort_ver21_hadm.columns])

df_merged = pd.merge(cohort_ver21_hadm, df_admissions, on='hadm_id', how='left')

# 시간 컬럼 형식 변환
time_cols = ['admittime', 'dischtime', 'deathtime']
for col in time_cols:
    df_merged[col] = pd.to_datetime(df_merged[col], errors='coerce')

# [1] 병원 재원 일수 los_days 계산
df_merged['los_days'] = (df_merged['dischtime'] - df_merged['admittime']).dt.total_seconds() / (60 * 60 * 24)

print("\n보완 1 los_days 계산 결과 is_icu_stay 별")
if 'is_icu_stay' in df_merged.columns:
    print(df_merged.groupby('is_icu_stay')['los_days'].describe())
else:
    print("is_icu_stay 컬럼 미존재")

# [2] 사망 시점 hours_to_death 재계산
# 우선순위 deathtime > dischtime > dod
if 'dod' in df_merged.columns:
    df_merged['dod'] = pd.to_datetime(df_merged['dod'], errors='coerce')
    df_merged['death_time_final'] = df_merged['deathtime'].fillna(df_merged['dischtime']).fillna(df_merged['dod'])
else:
    df_merged['death_time_final'] = df_merged['deathtime'].fillna(df_merged['dischtime'])

df_merged['hours_to_death'] = (df_merged['death_time_final'] - df_merged['admittime']).dt.total_seconds() / 3600
```

### (2) 시술 중 사망하거나 수술실로 직행해서 사망했는지 확인

```python

# 일반 병동 Non-ICU 사망자 상세 확인
if 'is_icu_stay' in df_merged.columns:
    non_icu_deaths = df_merged[
        (df_merged['is_icu_stay'] == 0) &
        ((df_merged['death_flag'] == 1) | (df_merged['disposition'] == 'EXPIRED'))
    ].copy()

    print(f"\n일반 병동 Non-ICU 사망: {len(non_icu_deaths)}명")

    if not non_icu_deaths.empty:
        # 시술 시도 후 24시간 내 사망자
        cath_deaths = non_icu_deaths[
            (non_icu_deaths['door_to_cath_min'].notna()) &
            (non_icu_deaths['hours_to_death'] < 24)
        ]

        print(f"\nCath 시도 후 24시간 내 사망: {len(cath_deaths)}명")

        if len(cath_deaths) > 0:
            print("\n상세 시술 중 사망 의심 환자")
            print(cath_deaths[['subject_id', 'door_to_cath_min', 'hours_to_death']])
    else:
        print("일반 병동 사망자 없음")
```

```python
import numpy as np
import pandas as pd
import sqlite3

# 1. DB 연결 설정
db_path = '/content/drive/MyDrive/Colab Notebooks/2025-2 데이터애널리틱스/MIMIC4-hosp-icu.db'

# 2. Admissions 테이블에서 병원 입/퇴원 시간 가져오기
try:
    conn = sqlite3.connect(db_path)
    # admittime(입원), dischtime(퇴원), deathtime(사망) 가져오기
    query = "SELECT hadm_id, admittime, dischtime, deathtime FROM admissions"
    df_admissions = pd.read_sql_query(query, conn)
    conn.close()
    print(f"✅ admissions 테이블 로드 완료 (Shape: {df_admissions.shape})")

    # hadm_id 타입 통일
    df_admissions['hadm_id'] = df_admissions['hadm_id'].astype('Int64')

    # 시간 형식 변환
    for col in ['admittime', 'dischtime', 'deathtime']:
        df_admissions[col] = pd.to_datetime(df_admissions[col], errors='coerce')

except Exception as e:
    print(f"⚠️ DB 연결 실패: {e}")

# 3. 데이터 병합 (cohort_ver22 + admissions)
# 방금 만드신 cohort_ver22를 사용합니다.
df_final = cohort_ver22.copy()
df_final['hadm_id'] = df_final['hadm_id'].astype('Int64')

# 혹시 기존에 잘못된 admittime 관련 컬럼이 있다면 제거하고 새로 붙입니다.
cols_to_drop = ['admittime', 'dischtime', 'deathtime', 'los_days']
df_final = df_final.drop(columns=[c for c in cols_to_drop if c in df_final.columns])

# 병합 (Left Join)
df_final = pd.merge(df_final, df_admissions, on='hadm_id', how='left')

# 4. 병원 재원 일수 (los_days) 및 사망 시간 재계산
# (1) 병원 재원 일수 (퇴원 - 입원)
df_final['los_days'] = (df_final['dischtime'] - df_final['admittime']).dt.total_seconds() / (60 * 60 * 24)

# (2) 입원 후 사망까지 시간 (hours_to_death) 업데이트
# admittime이 새로 들어왔으므로 정확도를 위해 업데이트
# death_time_final이 이미 있다면 사용하고, 없으면 재정의
if 'death_time_final' not in df_final.columns:
    if 'dod' in df_final.columns:
        df_final['dod'] = pd.to_datetime(df_final['dod'])
        df_final['death_time_final'] = df_final['deathtime'].fillna(df_final['dischtime']).fillna(df_final['dod'])
    else:
        df_final['death_time_final'] = df_final['deathtime'].fillna(df_final['dischtime'])

df_final['hours_to_death'] = (df_final['death_time_final'] - df_final['admittime']).dt.total_seconds() / 3600

# 5. 컬럼 정리 및 확인
print("\n✅ 최종 병합 및 계산 완료")
print(f"총 컬럼 수: {len(df_final.columns)}")

# 확인: admittime과 los_days가 있는지 확인
check_cols = ['admittime', 'dischtime', 'los_days', 'icu_los_days']
print(f"필수 컬럼 존재 여부: {[c for c in check_cols if c in df_final.columns]}")
```
