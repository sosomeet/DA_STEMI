```python
import pandas as pd

# 1) ver182 불러오기
path182 = "/content/drive/MyDrive/Colab Notebooks/cohort_ver182_winsor_99p.csv"
df = pd.read_csv(path182)

print("ver182 shape:", df.shape)
df.head(3)
```

```python
# 2) delay 컬럼 리스트
delay_cols = ["door_to_ecg", "door_to_trop", "door_to_anti", "door_to_pci"]

# 3) 각 delay에 대해 missing flag 생성
for col in delay_cols:
    flag_col = col + "_missing"
    df[flag_col] = (df[col] == -1).astype(int)

# 4) 잘 만들어졌는지 확인
print("===== 새로 추가된 missing flag 컬럼들 =====")
print([c for c in df.columns if "missing" in c])

print("\n===== 예시 5행 =====")
print(df[delay_cols + [c + "_missing" for c in delay_cols]].head())
```

```python
for col in delay_cols:
    flag_col = col + "_missing"
    print(f"\n=== {col} ===")
    print(df[[col, flag_col]].head())
    print("missing(1) 비율:", df[flag_col].mean())
```

```python
# 4) 새로운 버전으로 저장
path183 = "/content/drive/MyDrive/Colab Notebooks/cohort_ver183_add_delay_missing_flag.csv"
df.to_csv(path183, index=False)

print("[INFO] 저장 완료:", path183)
print("최종 shape:", df.shape)
```
